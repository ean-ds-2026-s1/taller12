### Ejercicio 05: promedio de una lista de noditos. Recursión directa.

Escriba un método llamado `promedio_tam_verbos` que reciba como dato de entrada la cabeza
de una lista de noditos de palabras lamada `cab` y que, usando una función auxiliar que
retorne una pareja (Un objeto de la clase `Pair`, el primer elemento es la suma, y el
segundo es el contador), retorne el promedio del tamaño (el número de letras) de aquellas
palabras de la lista de noditos que sean verbos terminados en minúsculas.
Use recursión directa para la función
auxiliar. **POR FAVOR: User las parejas, es un requerimiento del ejercicio. Usar tuplas
o cualquier estructura de datos de Python, no solo está prohibido sino que hará que
obtenga una mala nota en todo el taller.**

```python
from ean import *

# Definición de la función de promedio del tamaño de verbos
def promedio_tam_verbos(cab : Nodito[str] | Nulo) -> float:

    # Definición de la función auxiliar que retorna una pareja.
    # El primer elemento debe ser la suma, y el segundo el contador
    def promedio_tam_verbos_aux(actual: Nodito[str] | Nulo) -> Pair[int, int]:
        pass

    # ---- Operacion principal ----
    res = promedio_tam_verbos_aux(actual=cab)
    return 0 if res.contador == 0 else res.suma / res.contador
```

---
Prueba de la función de contar_verbos con recursión directa

```python
import unittest
from ean import *

class TestEjercicio05(unittest.TestCase):
    def test_ejercicio05(self):
        pals = crear_lista_nodos_palabras("comer", "ando", "andar", "partir", "suma",
                           "sumar", "reir", "ir", "dono", "poder")
        self.assertAlmostEqual(4.5714, promedio_tam_verbos(pals), delta=0.0001)

        pals = crear_lista_nodos_palabras("amor", "comi", "papa", "guerra")
        self.assertEqual(0.0, promedio_tam_verbos(pals))

        pals = crear_lista_nodos_palabras("AMAR", "comer", "almorzar", "TEMER", "sufre", "sufrir")
        self.assertAlmostEqual(6.3333, promedio_tam_verbos(pals), delta=0.0001)

        print("Prueba superada 👍")

# -------------------------------------------------------------------------
unittest.main(argv=['', '-v', 'TestEjercicio05'], verbosity=2, exit=False);
```
